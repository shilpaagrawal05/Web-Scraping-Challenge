```python
# URL of page to be scraped
url = 'https://mars.nasa.gov/news/'

# Retrieve page with the requests module
browser.visit(url)
html = browser.html

# Create BeautifulSoup object; parse with 'lxml'
soup = BeautifulSoup(html, 'html.parser')
```

```python
news_title = soup.find('div',class_ = 'list_text').find('div',class_= 'content_title')
print(news_title.text.strip())
news_paragraph = soup.find('div',class_ = 'list_text').find('div',class_= 'article_teaser_body')
print(news_paragraph.text.strip())
```

```python
url = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
# results = soup.find_all('div',class_ = 'collapsible results')
# for result in results:
#     title = result.find('div',class_ = 'description').find('h3').text
#     print(title)
```

```python
results = soup.find_all('div',class_ = 'description')
title = []
for result in results:
    h3 = result.find('h3').text
    title.append(h3)
    
print(title)   
    
```

```python
results = soup.find_all('img',class_ = 'thumb')
image_url = []
for result in results:
    url = 'https://astrogeology.usgs.gov' + result['src']
    image_url.append(url)
print(image_url)
```

```python
hemisphere_image_url = []
for x in range(4):
    dict = {'title':title[x],'image_url':image_url[x]}
    hemisphere_image_url.append(dict)    
hemisphere_image_url
```
